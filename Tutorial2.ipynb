```python
# 2.From python dictionary
import pandas as pd
weather_data = {
    'day' : ['01-01-2017', '02-01-2017', '03-01-2017', '04-01-2017'],
    'temperature' : [32, 35, 29, 25],
    'windspeed' : [9,6,8,7],
    'event' : ['Rainy', 'Sunny', 'Snow', 'Sunny']
}
df = pd.DataFrame(weather_data)
df
```

```python
# 3.From list of tuples
import pandas as pd
weather_data = [
    ('01-01-2017', 35, 9, 'Rain'),
    ('02-01-2017', 30, 5, 'Snow'),
    ('03-01-2017', 27, 8, 'Sunny'),
    ('04-01-2017', 28, 7, 'Rain')
]
df = pd.DataFrame(weather_data, columns = ["day", "temperature", "windspeed", "events"])
df
```

```python
# 4.From list of dictionaries
#in this appraoch each elements in the list represent one row along with column specification 
weather_data = [
    {'day' : '01-02-2017', 'temperature' : 25, 'windspeed': 6, 'event': 'Rain'},
    {'day' : '02-02-2017', 'temperature' : 32, 'windspeed': 8, 'event': 'Sunny'},
    {'day' : '03-02-2017', 'temperature' : 28, 'windspeed': 7, 'event': 'Snow'},
]
df = pd.DataFrame(weather_data)
df
```

```python
#to replace "not availbale", "n.a" with "NaN"
#this is basically use in cleaning the messay data 
import pandas as pd
df = pd.read_csv(r'C:\Users\Johnny\OneDrive\Desktop\Excel\stock_data.csv', na_values=["not available", "n.a"])
df
```

```python
#we can see that in revenue  there is -1, revenue can never be -1. so inorder to convert -1 into NaN
#without converting "eps" -1 into NaN. We can use "dict"
import pandas as pd
df = pd.read_csv(r'C:\Users\Johnny\OneDrive\Desktop\Excel\stock_data.csv', na_values={
    'eps' : ["not available", "n.a"],
    'revenue' : ["not available", "n.a", -1],
    'people' : ["not available", "n.a"]
})
df
```

```python
# 2. Write csv
df.to_csv('C:\Users\Johnny\OneDrive\Desktop\Excel\writing_stock_data.csv', index=False)
```
