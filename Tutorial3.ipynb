```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Johnny\OneDrive\Desktop\Excel\missing_data.csv")
df.set_index("day", inplace = True)
df
```

```python
# 1. fillna to fill missing values using different ways
new_df = df.fillna(0)
new_df
#here all the missing values got converted into "0.0"
```

```python
#event column cannot be "0", since it is in str format, we are replacing "0" to "no event"
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'no event'
})
new_df
```

```python
#instead of keeping "0" in all the missing values, we can fill it with some other values, like the value from forward or backward values.
#df.fillna(method= "ffill")  filling the missing values with forward value
#df.fillna(method= "bfill")  filling the missing values with backward value

new_df = df.fillna(method = "ffill")
new_df
```

```python
new_df = df.fillna(method = "bfill")
new_df
```

```python
#or we can also use the horizontal values, i,e. values from same rows 
#axis="columns"
new_df = df.fillna(method= "bfill", axis = "columns")
new_df
```

```python
#limit method, it is used when there is missing values continuoiusly and you want to copy the forward values only up some limit in the next adjacent missing values.

new_df = df.fillna(method = "ffill", limit = 1)
new_df
```

```python
# 2. interpolate ,to make a guess on missing values using interpolation
#interpolate() method is used to guess the values and put it in missing values
new_df = df.interpolate()
new_df
```

```python
# 3. dropna to drop rows with missing values
new_df = df.dropna()
new_df
#so here it basically drops all the rows that has missing values in cells
```

```python
#to drop only the rows that has a missing values in all the cells of rows
#how(method)
new_df = df.dropna(how="all")
new_df
```

```python
#thresh(), by using thresh(process) we can decide, if we want to keep that particular rows or not. 
new_df = df.dropna(thresh = 1)
new_df
#keeping the thresh value "1" we kept the rows which has atleast 1 data in a cell


new_df = df.dropna(thresh = 2)
new_df
#likewise

#However, day is not counted, because it is index
```

```python
#HANDLINg MISSING DATA: replace function


import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\Johnny\OneDrive\Desktop\Excel\missing_data2.csv")
df
```

```python
#to replace the "-99999" with NaN
#use replace method

new_df = df.replace([-99999, -9999], np.NaN)
new_df
```

```python
new_df = df.replace([-99999, -9999,"no event"], np.NaN)
new_df
```

```python
#to replace only the "no event" to "Sunny"

new_df = df.replace({
    -99999:np.NaN,
    -9999: np.NaN,
    "no event" : "Sunny"
})
new_df
```
