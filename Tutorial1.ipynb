```python
# 4.Conditional selection
```

```python
# like SQL, dataframe also supports like kind of sql where we can select our condition data and execute.

df[df.Temperature>= 30]
```

```python
df[df.Temperature == df.Temperature.max()]
```

```python
#to print only the specific column where the temp was max

df['Day'] [df.Temperature == df.Temperature.max()]

df[['Day','Temperature']] [df.Temperature == df.Temperature.max()]
```
